```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("DQN.csv")
```

```python
data.columns = ['Episode', "Step", "Epsilon", "Reward", "Loss", "Last 100 reward"]
```

```python
data.head()
```

```python
plt.figure(figsize = (8,6))
plt.plot(data["Episode"],data["Reward"], color='tab:orange', label ='Reward')
plt.plot(data["Episode"],data["Last 100 reward"], color ='b', label='Last 100 Reward', linewidth=2.5)
plt.xlabel("No of Episodes")
plt.ylabel("Reward")
plt.title("Rewards and Last 100 Rewards vs No. of Episodes")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
# plt.show()
plt.savefig("Reward vs Episodes.png")
```

```python
plt.figure(figsize = (9,6))
plt.plot(data["Episode"],data["Loss"], color ="tab:cyan", label = "Loss")
plt.xlabel("No of Episodes")
plt.ylabel("Loss")
plt.title("Loss vs No. of Episodes")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
# plt.show()
plt.savefig("Loss vs Episodes.png")
```

```python
data4 = pd.read_csv("data4.csv")
```

```python
data4.head()
```

```python
data4.columns = ['Frame', "Epsilon", "Loss","Reward",  "Last 10 reward"]
```

```python
data4.head()
```

```python
reward = data4["Last 10 reward"]
```

```python
reward = [np.mean(reward[i-20:i]) for i in range(0, len(reward))]
```

```python
plt.figure(figsize = (8,6))
plt.plot(data4["Frame"],data4["Reward"], color='tab:orange', label ='Reward')
plt.plot(data4["Frame"],reward, color ='b', label='Last 10 Reward', linewidth=2.5)
plt.xlabel("No of Frames")
plt.ylabel("Rewards")
plt.title("Rewards and Last 10 Rewards vs No. of Frames")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
# plt.show()
plt.savefig("Reward vs Frames-4.png")
```

```python
plt.figure(figsize = (9,6))
plt.plot(data4["Frame"],data4["Loss"], color ="tab:cyan", label = "Loss")
plt.xlabel("No of Frames")
plt.ylabel("Loss")
plt.title("Loss vs No. of Frames")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
# plt.show()
plt.savefig("Loss vs Frames-2.png")
```
